```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


# Set up Chrome options
options = webdriver.ChromeOptions()
options.add_argument('--no-sandbox')
options.add_argument('--headless')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--lang=en-US')


# Initialize the Chrome driver
driver = webdriver.Chrome(options=options)

# List of cities to search
cities = ["toronto"]  # Example cities; adjust as needed



for city in cities:
    url = f'https://www.facebook.com/marketplace/{city.lower()}/vehicles/?sortBy=creation_time_descend&exact=false'
    print(f'Searching in {city}: {url}')
    driver.get(url)
    language = driver.execute_script("return navigator.language")
    print(f"The browser language is: {language}")
    page_source = driver.page_source
    print(page_source)  # This will display the HTML of the loaded page


```

```python

try:
    # Wait for the close button to become clickable
    close_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//div[@aria-label="Close" and @role="button"]'))
    )
    close_button.click()
    print("Decline optional cookies button clicked!")
except Exception as e:
    print("Could not find or click the close button!", e)
    pass

# Close the driver

```

```python
try:
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
        print("scrolled")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
html = driver.page_source


soup = BeautifulSoup(html, 'html.parser')


driver.close()
```

```python
# Assuming 'soup' is already defined and contains the HTML of the Facebook Marketplace page
links = soup.find_all('a')


product_data = []

# Filter for links that match the category
product_links = [link for link in links ]

# Collect relevant data from each product link
for product_link in product_links:
    url = product_link.get('href')  # Get the URL of the product
    text = '\n'.join(product_link.stripped_strings)  # Get the text associated with the link
    product_data.append({'text': text, 'url': url})
print(product_data)
```
